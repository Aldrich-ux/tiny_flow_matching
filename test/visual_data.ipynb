# Visual Synthetic Data

Remember a kind of *Dataset* is just a *Distribution*, so data here is a dataset with 50000 samples/points. You could call it *Checkerboard Dataset*, but I refer to call it *Checkerboard Distribution*. 

```python
from dataset.synthetic_dataset import DatasetCheckerboard

dataset = DatasetCheckerboard()
data = dataset.sample(n=50000) # [n, 2]
```

```python
import matplotlib.pyplot as plt

x1 = data.clone()

plt.figure(figsize=(6, 4))
plt.hist2d(x1[:, 0], x1[:, 1], bins=300, cmap='viridis')
plt.colorbar(label='Density')
plt.title('Checkboard Data Density (viridis)')
plt.gca().set_aspect('equal', 'box')
```

```python
# `x_0` refer to Gaussian Distribution samples.
x0 = torch.randn(size=(50000, 2))

plt.figure(figsize=(6, 4))
plt.hist2d(x0[:, 0], x0[:, 1], bins=300, cmap='viridis')
plt.colorbar(label='Density')
plt.title('Checkboard Data Density (viridis)')
plt.gca().set_aspect('equal', 'box')
```

```python
t = 0.6 # 0 ~ 1
x_t = (1 - t) * x0 + t * x1

plt.figure(figsize=(6, 4))
plt.hist2d(x_t[:, 0], x_t[:, 1], bins=300, cmap='viridis')
plt.colorbar(label='Density')
plt.title('Checkboard Data Density (viridis)')
plt.gca().set_aspect('equal', 'box')
```
